```python
import pandas as pd 

vig1 = pd.read_excel('39.xlsx')
vig2 = pd.read_excel('16.xlsx')
```

```python
from bs4 import BeautifulSoup

import json
saver = {}
for row in range(len(vig1)):
    soup = BeautifulSoup(vig1.loc[row]['Ответ'], 'lxml')
    
    if vig1.loc[row]['Тип ответа'] not in saver.keys():
        saver[vig1.loc[row]['Тип ответа']] = {row: {
        'Вопрос': vig1.loc[row]['Название'],
        'Ответ': soup.text
    }}
    else:
         saver[vig1.loc[row]['Тип ответа']][row] = {
        'Вопрос': vig1.loc[row]['Название'],
        'Ответ': soup.text
    }
    
        

with open('data.json', 'w') as f:
    json.dump(saver, f)
```
